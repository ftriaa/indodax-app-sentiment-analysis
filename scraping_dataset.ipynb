#**Identifiksi Data**
Dataset yang akan di scraping ialah ulasan pengguna aplikasi Indodax dari Google Play Store dengan laman https://play.google.com/store/apps/details?id=id.co.bitcoin . Dataset ini hanya akan diambil 3 kolom utama yang akan digunakan, yaitu content, score dari 1-5, dan tanggal ulasan itu diunggah.

#**Import Library**

```python
from google_play_scraper import reviews, Sort, app
import pandas as pd
from google.colab import files
```

#**Scraping Data**

```python
result, _ = reviews(
    'id.co.bitcoin',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=50000,
    filter_score_with=None
)
```

#**Loading Dataset**

```python
df = pd.DataFrame(result)
df.head()
```

```python
df.info()
```

```python
df_cleaned = df.dropna()
df_cleaned = df_cleaned.drop_duplicates()
df_cleaned.info()
```

```python
df_cleaned.to_csv('indodax_reviews.csv', index=False, encoding='utf-8-sig')
files.download('indodax_reviews.csv')
```
